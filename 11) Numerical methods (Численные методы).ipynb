В общей сумме мы имеем 5 признаков с пропусками, и все являются важными.

**VehicleType:**

Проверим какое количество каждого типа мы имеем в данном признаке:

```python
data['VehicleType'].value_counts(dropna=False)
```

В данном случае мы видим, что данные распределяются по типам кузова у автомобилей. Так же имеем большое количество пустых значений (более 10%), которые невозможно восстановить. Их нельзя исскуственно приравнять или найти медианное значение. Неопределенные типы кузовов можно слить с категорией "другие", так как они действительно могут не попадать под классификацию. 

```python
data['VehicleType'] = data['VehicleType'].fillna('other')
```

```python
data['VehicleType'].value_counts(dropna=False)
```

**Gearbox:**

Проверим какое количество каждого типа мы имеем в данном признаке:

```python
data['Gearbox'].value_counts(dropna=False)
```

В данном случае мы видим доминирование механики на рынке, и почти в 4 раза меньше автомат. Так же в общую сумму входят позиции с пропусками. Тут нужно учесть два момента. Во первых доля пропусков достаточно велика (более 5%) и удалить данные пропуски без искажений мы не можем. Так же нельзя восстановить эти данные, так как несмотря на конкретные модели автомобилей, на них могут быть установлены как механическая, так и автоматическая коробки. Во вторых, помимо механики и автоматики, на автомобили могут ставить другие виды коробок, которые могут входить в это общее количество. Поэтому логичным выходом из этой ситуации, все пропуски так же переместить в категорию "другое", чтобы оставить пласт информации в целом и не искажать данные. 

```python
data['Gearbox'] = data['Gearbox'].fillna('other')
```

```python
data['Gearbox'].value_counts(dropna=False)
```

**Model:**

В отличии от прошлых признаков, данный является (если это можно так назвать) "корневым" который определяет все остальные признаки, так как именно от ТТХ конкретной модели строятся все параметры. И поэтому восстановить данные нельзя в принципе. В данном случае можно либо полностью удалить строки, или создать очередную категорию "другое", и слить туда все пропущенные позиции. 

```python
data['Model'] = data['Model'].fillna('other')
```

**FuelType:**

```python
data['FuelType'].value_counts(dropna=False)
```

Тут мы видим похожую ситуацию с большим количеством пропусков. Так же имеем категорию "разное", что подразумевает некоторую долю неизвестного или неподходящего вида топлива. В данной ситуации поступим аналогично и переведем пропуски в категорию "другое".

```python
data['FuelType'] = data['FuelType'].fillna('other')
```

В данном признаке есть только две классификации - это положительный и отрицательный ответ. Пропуски же могут означать как отрицательный ответ, так и нежелание заполнить данную графу пользователем. В любом случае мы не можем искусственно заполнить ее без сильных искажений данных, а тем более не можем удалить их. Но тем не менее можно пропуски перевести в категорию отсутсвия ремонта, что в теории возможно, и как минимум часть информации будет правдива.
```python
cod_set = ['VehicleType', 'Gearbox', 'Model', 'FuelType', 'Brand', 'Repaired']
```

```python
target = data_mod['Price']
features = data_mod.drop('Price', axis=1)
```

Базовые константы:

```python
test_size = 0.2
random_state = 12345
```

```python
features, features_test, target, target_test = train_test_split(features, target, test_size=test_size, random_state=random_state)
features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=test_size, random_state=random_state)
```

**Кодирование features_train:** 

```python
features_train_ohe = features_train
features_valid_ohe = features_valid
features_test_ohe = features_test
```

```python
ohe_encoder = OneHotEncoder(handle_unknown='ignore', drop='first', sparse=False)
```

```python
train_temp = ohe_encoder.fit_transform(features_train_ohe[cod_set])
features_train_ohe[ohe_encoder.get_feature_names()] = train_temp
features_train_ohe.drop(cod_set, axis=1, inplace=True)
```

**Кодирование features_valid:** 

```python
valid_temp = ohe_encoder.transform(features_valid_ohe[cod_set])
features_valid_ohe[ohe_encoder.get_feature_names()] = valid_temp
features_valid_ohe.drop(cod_set, axis=1, inplace=True)
```

**Кодирование features_test:** 

```python
test_temp = ohe_encoder.transform(features_test_ohe[cod_set])
features_test_ohe[ohe_encoder.get_feature_names()] = test_temp
features_test_ohe.drop(cod_set, axis=1, inplace=True)
```

```python
numeric = ['RegistrationYear', 'Power', 'Kilometer']
scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train_ohe[numeric] = scaler.transform(features_train_ohe[numeric])
features_test_ohe[numeric] = scaler.transform(features_test_ohe[numeric])
features_valid_ohe[numeric] = scaler.transform(features_valid_ohe[numeric])    
```

```python
%%time
model_linear = LinearRegression()
model_linear.fit(features_train_ohe, target_train)
```

```python
%%time
predictions_valid_linear = model_linear.predict(features_valid_ohe)
rmse = mean_squared_error(target_valid, predictions_valid_linear)**0.5
print('Метрики по валидной модели')
print('RMSE:', rmse)
```

**Вывод:** Скорость обучения линейной регрессии равна **1.74** секунды, скорость предсказания **55** миллисекунды при RMSE: **2624**
