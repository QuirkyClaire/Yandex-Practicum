## Подготовка данных

### Описание данных

Технологический процесс:

- **Rougher feed** — исходное сырье
- **Rougher additions** (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- **Xanthate** — ксантогенат (промотер, или активатор флотации);
- **Sulphate** — сульфат (на данном производстве сульфид натрия);
- **Depressant** — депрессант (силикат натрия).
- **Rougher process** (англ. «грубый процесс») — флотация
- **Rougher tails** — отвальные хвосты
- **Float banks** — флотационная установка
- **Cleaner process** — очистка
- **Rougher Au** — черновой концентрат золота
- **Final Au** — финальный концентрат золота

Параметры этапов:

- **air amount** — объём воздуха
- **fluid levels** — уровень жидкости
- **feed size** — размер гранул сырья
- **feed rate** — скорость подачи

Наименование признаков:

Возможные значения для блока [этап]:
- **rougher** — флотация
- **primary_cleaner** — первичная очистка
- **secondary_cleaner** — вторичная очистка
- **final** — финальные характеристики

Возможные значения для блока [тип_параметра]:
- **input** — параметры сырья
- **output** — параметры продукта
- **state** — параметры, характеризующие текущее состояние этапа
- **calculation** — расчётные характеристики

### Проверка формулы вычисления эффективности обогащения

**Вычислить MAE и обогащение (Recovery) на обучающей выборке для признака rougher.output.recovery:**

Определим формулу расчета для вычислений:

**Recovery = ((С * (F - T)) / (F * (C - T))) * 100**, где:

- **С** - доля золота в концентрате после флотации/очистки;
- **F** - доля золота в сырье/концентрате до флотации/очистки;
- **T** - доля золота в отвальных хвостах после флотации/очистки.

Определим требуемые признаки для вычислений:

- **rougher.output.recovery** - Доля "грубой" обработки (исследуемый признак).
- **rougher.output.concentrate_au** - концентрация золота после процесса флотации (С).
- **rougher.input.feed_au** - концентрация золота до процесса флотации (F).
- **rougher.output.tail_au** - концентрация золота в отвальных "хвостах" после флотации (T).

Расчитаем по формуле данные параметры для нахождения искомой величины. Создадим новый датасет из данных признаков для удобства, и возможного вызова в будущем. 

```python
recovery_data_train = ['rougher.output.recovery', 'rougher.output.concentrate_au', 'rougher.input.feed_au', 'rougher.output.tail_au']
recovery_data_train = data_train[recovery_data_train]
```

```python
recovery_data_train.info()
```

```python
C = recovery_data_train['rougher.output.concentrate_au']
F = recovery_data_train['rougher.input.feed_au'] 
T = recovery_data_train['rougher.output.tail_au']
```

```python
recovery_data_train['recovery_data_train'] = ((C * (F - T)) / (F * (C - T))) * 100
```

```python
recovery_data_train.head()
```

Расчитаем МАЕ:

```python
mae_recovery_data_train = mean_absolute_error(recovery_data_train['rougher.output.recovery'], recovery_data_train['recovery_data_train'])
```

```python
print('Среднее значение Recovery', recovery_data_train['recovery_data_train'].mean())
print('Значение MAE', mae_recovery_data_train)
```

**Вывод:** Значение MAE получили очень низкое, близкое к нулю. Это говорит о том, что значения recovery очень точные.

В файлах отсутствует большая часть информации - около 22%, отчего просто удалить мы ее не можем. Также приведение к медианным значениям будет спорным решением, так как значения в позициях очень похожи и замена на медиану может исказить результаты. Можно использовать метод ffill, который позволит наполнить пропуски если не точными, но максимально разнообразными значениями.

Функция для метода ffill:

```python
def data_ffill(data_file):
    data_file = data_file.fillna(method = 'ffill')
    return data_file
```

### Функция для расчета sMAPE:

Функция sMAPE:

```python
def smape (target, predictions):
    numerator = abs(predictions - target)
    denominator = (abs(target) + abs(predictions)) / 2
    summa = (numerator / denominator).sum()
    smape = (1 / len(target)) * summa * 100
    return smape
```

Функция расчета общего sMAPE:

```python
def ml(model):
    smape_rougher = cross_val (model, train_features_rougher, train_target_rougher)
    smape_final = cross_val (model, train_features_final, train_target_final)
    smape_end = 0.25 * smape_rougher + 0.75 * smape_final
    print('Средняя оценка качества по кросс-валидации на этапе флотации составила', smape_rougher)
    print('Средняя оценка качества по кросс-валидации на финальном этапе составила', smape_final)
    print('Конечная оценка качества по кросс-валидации составила', smape_end)
```

**Выбор подходящей модели для обучения:**

Функция оценки с помощью кросс-валидации:

```python
def cross_val (model, features, target):
    scores = cross_val_score(model, features, target, scoring=make_scorer(smape, greater_is_better=True), cv=5,n_jobs=-1)
    smape_res = np.mean(scores)
    return smape_res
```

Функция расчета общего sMAPE:

```python
def ml(model):
    smape_rougher = cross_val (model, train_features_rougher, train_target_rougher)
    smape_final = cross_val (model, train_features_final, train_target_final)
    smape_end = 0.25 * smape_rougher + 0.75 * smape_final
    print('Средняя оценка качества по кросс-валидации на этапе флотации составила', smape_rougher)
    print('Средняя оценка качества по кросс-валидации на финальном этапе составила', smape_final)
    print('Конечная оценка качества по кросс-валидации составила', smape_end)
```

**Подбор более оптимальных гиперпараметров для модели:**

```python
best_smape = 13
best_n_estimators = 0
best_depth = 0
for n_estimators in tqdm(range(1, 10)):
    for max_depth in range (2, 6):
        for min_samples_split in range(2, 6):
            for min_samples_leaf in range(2, 6):
                model = RandomForestRegressor(random_state=12345, n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)
                smape_rougher = cross_val (model, train_features_rougher, train_target_rougher)
                smape_final = cross_val (model, train_features_final, train_target_final)
                smape_end = 0.25 * smape_rougher + 0.75 * smape_final
                if smape_end < best_smape:
                    best_smape = smape_end
                    best_depth = max_depth
                    best_parameter = [n_estimators, best_depth, min_samples_split, min_samples_leaf]
print('Гиперпараметры модели:')
print('sMAPE', best_smape)
print('n_estimators:', best_parameter[0]) 
print('max_depth:', best_parameter[1]) 
print('min_samples_split:', best_parameter[2])
print('min_samples_leaf:', best_parameter[3])      
```

**Вывод:** Мы получили оптимальные гиперпараметры для модели рандомного леса: **n_estimators: 6**, **depth - 5**, **n_samples_split - 2** и **min_samples_leaf - 5**. Значение sMAPE при этом **10.29%**.
