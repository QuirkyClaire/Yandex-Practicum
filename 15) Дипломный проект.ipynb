<h1>Проект: Промышленность "ООО «Так закаляем сталь»"</h1>

# Описание:

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. 

**Производственный процесс:**\
Комбинат ООО «Так закаляем сталь» производит большой ассортимент марок стали с различными свойствами, в основном это прочность, гибкость, устойчивость к окружающей среде и т.п. Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Производственный процесс делится на этапы, в каждый из которых добавляются дополнительные смеси и сплавы в общую массу. После каждой итерации, смесь продувается кислородом или другим инертным газом и нагревается подготавливая сплав к следующей итерации. Каждая марка имеет свой состав присадок и твердотельных сплавов, которые входят в ее состав. От количества составляющих зависит количество итераций и нагревов сплава. Согласно ТЗ, нужно провести анализ данных входных данных и предсказать какая температура стали будет на выходе. 


**Задачи:** 
- Обучить модель, которая предскажет температуру стали;
- Добиться значение метрики MAE не более 6,8.

**Описание данных:**
- data_arc_new.csv — данные об электродах;
- data_bulk_new.csv — данные о подаче сыпучих материалов (объём);
- data_bulk_time_new.csv — данные о подаче сыпучих материалов (время);
- data_gas_new.csv — данные о продувке сплава газом;
- data_temp_new.csv — результаты измерения температуры;
- data_wire_new.csv — данные о проволочных материалах (объём);
- data_wire_time_new.csv — данные о проволочных материалах (время).

**План работы:**

- Загрузим данные из предоставленных файлов и изучим их
- Сделать выводы по файлам и сформировать направление анализа для каждого из них
- Сделать общий вывод по данным датасетам
- Провести анализ каждого датасета и сформировать вывод
- Провести предобработку датасетов.
- Объединить все датасеты в одну таблицу и убрать лишние признаки. Определить целевой признак.
- Выбрать модели для обучения и найти лучшую для выполнения задачи.
- Провести тестирование модели.
- Сформировать вывод

**Создание признака "полной мощности":**

Чтобы получить полную мощность имея активную и реактивную мощности, нужно использовать формулу "треугольник мощностей", который позволит высчитать значение полной мощности. 

$S = \sqrt[]{P^2+Q^2}$

где: 
- Q - реактивная мощность
- P - активная мощность
- S - полная мощность

```python
data_arc_new['Полная мощность'] = ((data_arc_new['Активная мощность']**2) + (data_arc_new['Реактивная мощность']**2)**0.5)
```

```python
cv = TimeSeriesSplit(n_splits=5).split(features_train)
model_lr = LinearRegression()
model_lr.fit(features_train, target_train)
score_lr = cross_val_score(model_lr, features_train, target_train, scoring='neg_mean_absolute_error', cv=cv)
print("MAE обучающей выборки:", round(np.mean(-score_lr), 2))
```

### CatBoost

```python
features = data.drop(['Последний замер'], axis=1)
target = data['Последний замер']
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state = 310723, shuffle=False, test_size=0.25)
```
