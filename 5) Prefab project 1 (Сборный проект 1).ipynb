### Преобразовать тип данных в нужные типы:

В таблице было найдено три столбца, имеющие некорректную типизацию данных, которые нужно исправить:
- Тип даты релиза нужно изменить с float на int, чтобы получить целочисленные значения. Но тут стоит один момент, нужно что-то делать с пропусками, которые не дают перевести данные в столбце в другой формат;
- Тип оценки пользователей также нужно перевести с object на float64. Задание намекает, что стоит порыться в данных столбика, и найти ошибочные позиции, которые вызывают ошибку в переводе.

В остальном таблица более не нуждается в манипуляциях с типами данных.

```python
data['year_of_release'].unique()
```

Для перевода столбца в другой формат нам мешают пропущенные значения nan, с которыми что-то нужно сделать. Посчитаем сколько вообще подобных значений в целом. (Комментарий: Тут есть один неочевидный момент, который я подглядел у других. Если считать общее количество пропусков, их становится недопустимо много - почти 2%. И становится не совсем понятно что делать дальше, так как закрыть глаза на это будет неправильно. Нижняя группировка меня удивила, особенно когда я понял что игры могут дублироваться на платформах, а значит их можно искусственно привести к +/- правильному значению.) Сгруппируем данные, чтобы попытаться исключить повторения:

```python
data = data.fillna({'year_of_release': data.groupby('name')['year_of_release'].transform('first')})
```

```python
print('Пропуски с датой:', len(data.loc[data['year_of_release'].isna()]))
print('Процент пропусков с датой:', len(data.loc[data['year_of_release'].isna()])/len(data)*100)
```

Мы смогли почти в половину сократить количество пропусков в столбце с датой. Теперь уровень потерь составляет 0,87 процентов, что допустимо для обобщения. Последний год в нашей таблице это 2016, и если взять для пропущенных значений медиану всего сетапа, можно получить примерные цифры для пропусков. Учитывая что это должен быть район 2000-х годов, в который постепенно начался рост в индустрии, будет минимальный ущерб для общей картины.

```python
data = data.dropna(subset=['year_of_release'])
```

```python
data['year_of_release'].unique()
```

Теперь наконец можно привести столбец с датами релиза к формату int:

```python
data['year_of_release'] = data['year_of_release'].astype(int)
```

```python
check_data = data['year_of_release']
```

```python
data.info()
```

## Проверка гипотез:

###  Средние пользовательские рейтинги платформ Xbox One и PC одинаковые:

**Проведем проверку гипотезы:**\
"Средние пользовательские рейтинги платформ Xbox One и PC одинаковые"\
**Формулировка нулевой гипотизы H0:**\
"Средняя пользовательская оценка игр на платформе Xbox One равна средней пользовательской оценке игр на платформе PC"\
**Формулировка альтернативной гипотезы H1:**\
"Средняя пользовательская оценка игр на платформе Xbox One не равна средней пользовательской оценке игр на платформе PC"\
Уровень значимости alpha = 0.05\
Выборки оценок по платформам XOne и PC не связаны друг с другом. Поэтому можно воспользоваться t - критерием Стьюдента.\

```python
data_ac = data[data['year_of_release'] >= 2012]
```

```python
from scipy import stats as st
```

```python
alpha = 0.05
```

```python
pc = data_ac[(data_ac['platform'] == 'PC') & (data_ac['user_score'] > 0)]['user_score']
xbox_one = data_ac[(data_ac['platform'] == 'Xbox One') & (data_ac['user_score'] > 0)]['user_score']
```

```python
results = st.ttest_ind(pc, xbox_one)
```

```python
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Итог:** Средняя пользовательская оценка игр на платформе Xbox One равна средней пользовательской оценке игр на платформе PC.

###  Средние пользовательские рейтинги жанров Action и Sports разные:

**Проведем проверку гипотезы:**\
"Средние пользовательские рейтинги жанров Action и Sports различаются"\
**Формулировка нулевой гипотизы H0:**\
"Средняя пользовательские рейтинги жанров Action совпадают с рейтингами Sports"\
**Формулировка альтернативной гипотезы H1:**\
"Средняя пользовательские рейтинги жанров Action не совпадают с рейтингами Sports"\
Уровень значимости alpha = 0.05\
Выборки рейтингов по жанрам Action и Sports не связаны друг с другом. Поэтому можно воспользоваться t - критерием Стьюдента.\

```python
alpha = 0.05
```

```python
action = data_ac[(data_ac['genre'] == 'Action') & (data_ac['user_score'] > 0)]['user_score']
sports = data_ac[(data_ac['genre'] == 'Sports') & (data_ac['user_score'] > 0)]['user_score']
```

```python
results1 = st.ttest_ind(action, sports)
```

```python
print('p-значение:', results1.pvalue)

if (results1.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Итог:** Средняя пользовательские рейтинги жанров Action и Sports различаются
